```python
import pandas as pd

tn = pd.read_csv("../data/processed/TN_processed.csv")
tp = pd.read_csv("../data/processed/TP_processed.csv")
nh3 = pd.read_csv("../data/processed/NH3_processed.csv")
no23 = pd.read_csv("../data/processed/N023_processed.csv")
op = pd.read_csv("../data/processed/OP_processed.csv")
ssc = pd.read_csv("../data/processed/SSC_processed.csv")

dfs = [tn, tp, nh3, no23, op, ssc]
for df in dfs:
    df["dateTime"] = pd.to_datetime(df["dateTime"])
    

```

```python
import pandas as pd

def load_csv(path, value_name):
    df = pd.read_csv(path)
    df["dateTime"] = pd.to_datetime(df["dateTime"])
    df.drop(columns=["code"], inplace=True, errors="ignore")   # drop code safely
    df.rename(columns={"value": value_name}, inplace=True)
    return df

tn  = load_csv("../data/processed/TN_processed.csv",  "TN")
tp  = load_csv("../data/processed/TP_processed.csv",  "TP")
nh3 = load_csv("../data/processed/NH3_processed.csv", "NH3")
no23= load_csv("../data/processed/N023_processed.csv","N023")
op  = load_csv("../data/processed/OP_processed.csv",  "OP")
ssc = load_csv("../data/processed/SSC_processed.csv", "SSC")

```

```python
final_df = tn.merge(tp, on="dateTime", how="outer")\
             .merge(nh3, on="dateTime", how="outer")\
             .merge(no23, on="dateTime", how="outer")\
             .merge(op, on="dateTime", how="outer")\
             .merge(ssc, on="dateTime", how="outer")

final_df.sort_values("dateTime", inplace=True)
final_df.reset_index(drop=True, inplace=True)

final_df.head()
final_df.info()

```

```python
final_df.fillna(method='ffill', inplace=True)
final_df.dropna(inplace=True)
final_df.info()
```

```python
final_df.to_csv("../data/final/master_dataset.csv", index=False)
```
